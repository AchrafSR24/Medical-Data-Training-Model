# About DataSet

The dataset includes a set of medical data concerning patients. Each entry is characterized by a unique identifier ('id') and a diagnosis ('diagnosis') indicating whether the patient has a Benign (B) or Malignant (M) tumor.  

In addition to this information, the dataset contains a series of numerical measurements associated with tumor characteristics, covering a comprehensive range of data for each patient, such as the mean radius ('radius_mean'), mean texture ('texture_mean'), mean perimeter ('perimeter_mean'), mean area ('area_mean'), smoothness ('smoothness_mean'), compactness ('compactness_mean'), etc.

***How we can say that this column is a non-relevant column?***
🚫 **1. Non-informative Columns (Identifiers, Metadata)**
These columns do not contain useful medical information:
✅ To be removed

id → A unique identifier, but it does not influence the diagnosis.
🔄 **2. Redundant Columns (Highly Correlated)**
Some columns are highly correlated, meaning they provide almost the same information. Example:

perimeter_mean and radius_mean are strongly related since the perimeter is proportional to the radius.
area_mean and radius_mean are also correlated (Area = π × Radius²).
✅ Solution: Keep only one of these variables to avoid redundancy.
📉 **3. Columns with Very Low Variance**
If a column contains nearly the same value all the time, it has little to no impact on the model. *Example:*
from sklearn.feature_selection import VarianceThreshold
selector = VarianceThreshold(threshold=0.01)  # Removes columns with very low variance
df_reduced = selector.fit_transform(df.drop(columns=['id', 'diagnosis']))
✅ Solution: Remove columns with extremely low variance.

🧪 **4. Noisy or Hard-to-Interpret Columns**
Some columns, such as fractal_dimension_mean, may be difficult to leverage in practice. If they do not improve the model, they can be ignored after analysis.

✅ Solution: Test the model with and without these variables to assess their impact.

**✅ Conclusion: Columns to Remove (Non-Relevant)**

**id → Identifier, not useful.**
**Redundant columns → Remove one of the highly correlated columns (perimeter_mean, area_mean, etc.).**
**Low-variance columns → Detect and remove those that do not contribute to the analysis.**
**Hard-to-interpret columns → Evaluate their impact on model accuracy before deciding.**

📌 En résumé :
✅ La normalité facilite les analyses, permet d’utiliser des méthodes paramétriques et aide à détecter les anomalies.

Si la distribution des données suit une loi normale, cela signifie que les valeurs sont réparties de manière symétrique autour de la moyenne, avec une concentration élevée au centre et une diminution progressive aux extrêmes.

📌 Conclusions et Implications d'une Distribution Normale
**🔹 1. Analyse statistique fiable**

Les méthodes statistiques classiques (moyenne, écart-type, etc.) sont pertinentes.
Les tests paramétriques (comme le t-test ou l’ANOVA) sont valides, car ils supposent souvent une normalité des données.
**🔹 2. Prédictibilité et prise de décision optimisée**

68% des données se trouvent dans [μ - σ, μ + σ].
95% des données se trouvent dans [μ - 2σ, μ + 2σ].
99.7% des données se trouvent dans [μ - 3σ, μ + 3σ].
👉 Cela aide à prédire la dispersion des données et à fixer des seuils de détection.
**🔹 3. Détection des valeurs aberrantes**

Une valeur en dehors de 3σ (±3 écarts-types de la moyenne) est rare et peut être une anomalie ou une donnée erronée.
**🔹 4. Simplification des Modèles de Machine Learning**

Beaucoup d'algorithmes de ML (comme régression linéaire, LDA, etc.) fonctionnent mieux avec des données normalement distribuées.
On peut normaliser ou standardiser les données (centrer-réduire) pour améliorer la performance des modèles.
**🔹 5. Utilisation dans la théorie des probabilités**

On peut utiliser la fonction de densité de probabilité (PDF) et la fonction de répartition (CDF) pour calculer la probabilité d’un événement.

```python
x=df.drop(columns=["diagnosis"])
y=df["diagnosis"]
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest= train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
tree=DecisionTreeClassifier()
tree.fit(xtrain, ytrain)
tree.score(xtest,ytest)
```

```python
plt.figure(figsize=(12, 8))  # Ajuster la taille du graphique
plot_tree(tree, filled=True)
plt.show()
```
